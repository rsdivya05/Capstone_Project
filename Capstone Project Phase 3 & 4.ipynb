```python
x=new_df.drop("Votes",axis=1)
y=new_df["Votes"]
```

```python
from sklearn.preprocessing import StandardScaler

scl= StandardScaler()
scaled_x = scl.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
Lr = LinearRegression()

max_accuracy = 0
max_random = 0

for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    Lr.fit(x_train,y_train)
    Lr_predict =Lr.predict(x_test)
    
    R2_score = r2_score(y_test,Lr_predict)
    if R2_score>max_accuracy:
        max_accuracy=R2_score
        max_random = i
        
print("R2 Score is - ", max_accuracy, "at the random state", max_random )
```

```python
Lr.fit(x_train, y_train)
Lr_predict = Lr.predict(x_test)

print("R2 Score :", r2_score(y_test, Lr_predict))
print("The Mean Absolute Error :", mean_absolute_error(y_test,Lr_predict))
print("The Mean Squared Error :", mean_squared_error(y_test,Lr_predict))

print("Training accuracy is -",Lr.score(x_train,y_train))
print("Testing accuracy is -",Lr.score(x_test,y_test))
```

```python
from sklearn.model_selection import cross_val_score

for c in range(2,10):
    CV = cross_val_score(Lr,x,y,cv=c)
    cv_mean= CV.mean()
    print(f'At cross fold {c} the cv score is {cv_mean}')
    print('\n')
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
Model1 = [DecisionTreeRegressor(), RandomForestRegressor(), KNeighborsRegressor(), AdaBoostRegressor(), GradientBoostingRegressor(), SVR()]
```

```python
for M in Model1:
    print("Accuracy of a Model   :", M)
    M.fit(x_train,y_train)
    predict_model = M.predict(x_test)
    print("R2 score     :", r2_score(y_test,predict_model)*100)
    
    print("MAE   :", mean_absolute_error(y_test,predict_model))
    
    print("MSE    :", mean_squared_error(y_test,predict_model))
    
    print("Training accuracy is -",M.score(x_train,y_train))
    print("Testing accuracy is -",M.score(x_test,y_test))
    
    model_score = cross_val_score(M,x,y, cv=7).mean()
    print("CV Score is    :", model_score)
    print("-----------------------------------------------------------------------------")
```

As a result, the RandomForestRegressor has the highest R2 score, which indicates a better fit to the data compared to other models. It also has relatively low MAE and MSE values indicates that it's capturing the variance in the data more accurately.
