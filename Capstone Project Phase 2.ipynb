```python
DB = sqlite3.connect("C:\\Users\\hp\\Capstone_Project.db")
cursor = DB.cursor()
```

```python
with open("C:\\Users\\hp\\Capstone_Project\\_IMDBDataset1_.csv",'r')as file:
    totalmovie =0
    content = csv.reader(file)
    next(content)  # Skip the header row if it exists

    for row in content:
        data_to_insert = tuple(row[:11])
        cursor.execute("INSERT INTO _IMDBDataset1_ (Serial_Number, Movie_name, Director_name, Director2, Director3, Director4, Director5, Duration, Year, Ratings, Metascore) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", data_to_insert)
        DB.commit()
        totalmovie += 1

print(totalmovie)
```

```python
query2 = cursor.execute("SELECT _IMDBDataset1_.Movie_name, Director_name, Director2, Director3, Ratings, _IMDBDataset2.Gross_Collection FROM _IMDBDataset1_ INNER JOIN _IMDBDataset2 ON _IMDBDataset1_.Movie_name = _IMDBDataset2.Movie_name WHERE Genre1='Action' OR Genre2='Action' OR Genre3='Action'")
query2.fetchall()
```

```python
query3 = cursor.execute("SELECT * FROM _IMDBDataset1_ INNER JOIN _IMDBDataset2 ON _IMDBDataset1_.Movie_name = _IMDBDataset2.Movie_name Where _IMDBDataset2.Gross_Collection= (SELECT MAX(_IMDBDataset2.Gross_Collection) FROM _IMDBDataset2 WHERE _IMDBDataset2.Gross_Collection!='NaN')")
query3.fetchall()
```

```python
query4 = cursor.execute("SELECT * FROM _IMDBDataset1_ JOIN _IMDBDataset2 ON _IMDBDataset1_.Movie_name = _IMDBDataset2.Movie_name WHERE _IMDBDataset1_.Ratings = (SELECT MAX(Ratings) FROM _IMDBDataset1_)")
query4.fetchall()
```

```python
# Query2 from Table 1 Dataframe

time_duration=imdb1[(imdb1["Duration"]>=140) & (imdb1["Duration"] <=190)]
print(len(time_duration))
time_duration
```

```python
# Query3 from Table 1 Dataframe

Rate = imdb1[imdb1["Ratings"] > 7].sort_values("Ratings", ascending=True)
print(len(Rate))
Rate
```

```python
# Query4 from Table 1 Dataframe

Movies = imdb1[['Movie_name']].sort_values(by="Movie_name", ascending=False)
print(len(Movies))
Movies
```

```python
# Query5 from Table 1 Dataframe

MOV_P = imdb1[imdb1['Movie_name'].str.startswith("P") & (imdb1['Ratings'] > 7)] [["Movie_name", "Ratings"]]
print(len(MOV_P))
MOV_P
```

```python
# Query1 from Table 2 Dataframe


arnold_movies = imdb2[imdb2['Stars'].str.contains("Arnold Schwarzenegger")]
arnold_movies = arnold_movies[["Movie_name", "Stars", "Star2", "Star3", "Star4"]]
arnold_movies_sorted = arnold_movies.sort_values("Movie_name", ascending=True)
arnold_movies_sorted
```

```python
# Query3 from Table 2 Dataframe

Grosscollection= imdb2[["Movie_name","Gross_Collection"]].sort_values("Gross_Collection",ascending = False)
print(len(Grosscollection))
Grosscollection
```

```python
# Query4 from Table 2 Dataframe

Arnold_gross= imdb2[imdb2['Stars'].str.contains("Arnold")]
Arnold_gross = Arnold_gross[["Movie_name", "Stars", "Star2", "Star3", "Star4", "Gross_Collection"]]
print(len(Arnold_gross))
Arnold_gross
```

```python
# Query1 from Both Table Dataframe

Merge_movies = pd.merge(imdb1,imdb2, on="Movie_name", how='inner')
print(len(Merge_movies))
Merge_movies
```
