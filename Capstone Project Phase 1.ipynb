```python
from bs4 import BeautifulSoup
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import re
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException,ElementNotInteractableException
from selenium.webdriver.support.ui import WebDriverWait

import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r"C:\web driver\chromedriver.exe")
```

```python
driver.get("https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1")
```

# We'll start with scraping data

```python
#file 1
Sno=[] 
Movie_name=[]
Director_name=[]
Duration=[]
year=[]
ratings=[]
metascore=[]

#file 2
votes=[]
stars=[]
Genre=[]
Gross_collection=[]
popularity=[]
Certification=[]
```

```python
# For File 1
start =0
end= 49
delay=10
for page in range(start,end):
    WebDriverWait(driver,delay)

    try:
        s_no = driver.find_elements(By.XPATH, '//span[@class="lister-item-index unbold text-primary"]')
        for i in s_no:
            Sno.append(i.text)
    except NoSuchElementException:
        Sno.append('-')
        
        
    try:
        movie = driver.find_elements(By.XPATH, '//h3[@class="lister-item-header"]')
        for i in movie:
            Movie_name.append(i.text.split()[1:-1])
    except NoSuchElementException:
        Movie_name.append('-')
        
    
    try:
        duration= driver.find_elements(By.XPATH, '//span[@class="runtime"]')
        for i in duration:
            Duration.append(i.text)
    except NoSuchElementException:
        Duration.append('-')
        
        
    try:
        Yr= driver.find_elements(By.XPATH, '//span[@class="lister-item-year text-muted unbold"]')
        for i in Yr:
            year.append(i.text)
    except NoSuchElementException:
        year.append('-')
        
        
    try:
        Rating= driver.find_elements(By.XPATH, '//div[@class="inline-block ratings-imdb-rating"]')
        for i in Rating:
            ratings.append(i.text)
    except NoSuchElementException:
        ratings.append('-')
            
    
    try:
        score= driver.find_elements(By.XPATH, '//div[@class="inline-block ratings-metascore"]')
        for i in score:
            metascore.append(i.text)
    except NoSuchElementException:
        metascore.append('None')
    
    
# For File 2
    try:
        Votes= driver.find_elements(By.XPATH, '//p[@class="sort-num_votes-visible"]')
        for i in Votes:
            votes.append(i.text.strip().split()[1])
    except NoSuchElementException:
        metascore.append('-')
        
    
    try:
        gross= driver.find_elements(By.XPATH, '//p[@class="sort-num_votes-visible"]')
        for i in gross:
            Gross_collection.append(i.text.strip().split()[-1])
    except NoSuchElementException:
        Gross_collection.append('-')
    
    try:
        genre= driver.find_elements(By.XPATH, '//span[@class="genre"]')
        for i in genre:
            Genre.append(i.text)
    except NoSuchElementException:
        Genre.append('-')

    try:
        certificate= driver.find_elements(By.XPATH, '//span[@class="certificate"]')
        for i in certificate:
            Certification.append(i.text)
    except NoSuchElementException:
        Certification.append('-')
    
    
# (For separating Director and Stars into two fields)
    
    try:
        link = driver.find_elements(By.XPATH, '//*[@id="main"]/div/div[3]/div/div[position()>=1 and position()<=1700]/div[3]/p[3]')
        for i in link:
            Director_name.append(i.text.strip()[10:].split('|')[0:1])
            stars.append(i.text.strip()[15:].split('|')[1:])
    except NoSuchElementException:
        Director_name.append('-')
        stars.append('-')

    
# Repeating this for all pages
    try:
        next_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
        next_button.click()
    except NoSuchElementException:
        pass
        time.sleep(3)
```

```python
driver.close()
```
